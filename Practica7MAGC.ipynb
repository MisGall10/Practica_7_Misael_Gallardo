* **Crear los siguientes graficos con Matplotlib**

a) Grafico de barras:
  * Objetivo: Comparar el promedio de largo y ancho de los pétalos entre las tres especies de Iris.
  * Instrucciones: Utiliza un gráfico de barras para mostrar la media de las columnas `petal_length` y `petal_width` agrupadas por `species`.

````python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
media = df.groupby('species')[['petal_length','petal_width']].mean().reset_index()
sns.barplot(data = media.melt(id_vars='species'), x = 'species', y = 'value',hue = 'variable', palette = 'Set2')
#Ponemos el titulo a el grafico
plt.title('Media de longitud y ancho del pétalo por especia', fontsize = 15, fontweight = 'bold')
#Ponemso el titulo del eje x
plt.xlabel('Especie', fontsize=10)
#Ponemos el titulo del eje z
plt.ylabel('Media de la longitud y ancho del pétalo por especia', fontsize=10)
#Imprimimos el resultado
plt.show()
````

b) Histogarama:
  * Objetivo: Visualizar la distribución de los largos de los sépalos (`sepal_length`).
  * Instrucciones: Crea un histograma de la variable `sepal_length` y asegúrate de que tenga un número adecuado de bins. Coloca etiquetas en los ejes y un título.

````python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
sns.histplot(df['sepal_length'], color = 'Royalblue', bins = 8, edgecolor = 'black', alpha = 0.4)
#Ponemos el titulo a el grafico
plt.title('Histograma de la longitud del sépalo', fontsize = 20)
#Ponemso el titulo del eje x
plt.xlabel('Longitud del sepalo (m)', fontsize=15)
#Ponemos el titulo del eje z
plt.ylabel('Frecuencia', fontsize=15)
#Imprimimos el resultado
plt.show()
````

c) Gráfico de dispersión (scatter plot):
  * Objetivo: Relacionar el largo del pétalo con el ancho del pétalo.
  * Instrucciones: Crea un gráfico de dispersión que compare `petal_length` con `petal_width`. Usa un color diferente para cada especie.

````python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
sns.scatterplot(data = df, x = 'petal_length',y = 'petal_width', hue = 'species', palette = 'Set3', s = 30, edgecolor = 'black')
#Ponemos el titulo a el grafico
plt.title('Relacion entre la longitud y el ancho del petalo', fontsize = 20)
#Ponemso el titulo del eje x
plt.xlabel('Longitud del pétalo (cm)', fontsize=15)
#Ponemos el titulo del eje z
plt.ylabel('Ancho del pétalo (cm)', fontsize=15)
#Imprimimos el resultado
plt.show()
````

d) Gráfico de cajas (box plot):
  * Objetivo: Analizar la distribución de las medidas del sépalo (`sepal_length`, `sepal_width`).
  * Instrucciones: Crea un gráfico de cajas para mostrar la distribución de las variables `sepal_length` y `sepal_width` por especie.

````python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
sns.boxplot(data = df, x = 'species',y = 'sepal_length', palette = 'Set1')
sns.boxplot(data = df, x = 'species',y = 'sepal_width', palette = 'Set2')
#Ponemos el titulo a el grafico
plt.title('Distribucion de longitud y ancho sépal por expecie', fontsize = 20)
#Ponemso el titulo del eje x
plt.xlabel('Especie', fontsize=15)
#Ponemos el titulo del eje z
plt.ylabel('Valor', fontsize=15)
#Imprimimos el resultado
plt.show()
````

e) Gráfico de líneas:
  * Objetivo: Mostrar la tendencia de la longitud del pétalo (`petal_length`) a través de las observaciones.
  *Instrucciones: Crea un gráfico de líneas utilizando la variable `petal_length` (en el eje Y) con el índice de las observaciones (en el eje X). Puedes hacer esto para ver cómo varía la longitud del pétalo a lo largo del dataset.

````python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
sns.lineplot(x = df.index,y = df['petal_length'], marker = 'o', color = 'green')
#Ponemos el titulo a el grafico
plt.title('Variacion de la longitud del pétalo a lo largo del Dataset', fontsize = 20)
#Ponemso el titulo del eje x
plt.xlabel('Indice de observaciones', fontsize=15)
#Ponemos el titulo del eje z
plt.ylabel('Longitud del pétalo (cm)', fontsize=15)
#Imprimimos el resultado
plt.show()
````
